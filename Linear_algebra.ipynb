Linear Algebra(vectors&metrices)

```python
import numpy as np

v = np.array([1, 2, 3])

A = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Vector:", v)
print("Matrix:\n", A)
```

Matrix Addition & Scalar Multiplication

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Addition:\n", A + B)
print("Scalar Multiplication:\n", 2 * A)
```

Matrix Multiplication

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Method 1
C = np.dot(A, B)

# Method 2 (recommended)
C2 = A @ B

print("Matrix Multiplication:\n", C2)
```

Transpose of Matrix

```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Transpose:\n", A.T)
```

Determinant

```python
A = np.array([[1, 2],
              [3, 4]])

det = np.linalg.det(A)
print("Determinant:", det)
```

Inverse of Matrix

```python
A = np.array([[1, 2],
              [3, 4]])

inv = np.linalg.inv(A)
print("Inverse:\n", inv)
```

Eigenvalues & Eigenvectors

```python
A = np.array([[2, 0],
              [0, 3]])

values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

Solving Linear Equations

```python
A = np.array([[2, 1],
              [1, 3]])

b = np.array([8, 13])

x = np.linalg.solve(A, b)

print("Solution x:", x)
```

Rank of Matrix

```python
A = np.array([[1, 2],
              [2, 4]])

rank = np.linalg.matrix_rank(A)
print("Rank:", rank)
```

Norm (Vector Length)

```python
v = np.array([3, 4])

norm = np.linalg.norm(v)
print("Norm:", norm)   # Should be 5
```
