N-Dimensional Array

```python
import numpy as np

arr = np.array([[1,2,3],
                [4,5,6]])

print(arr)
```

Number of Dimensions

```python
print(arr.ndim)
```

shapes


```python
print(arr.shape)
```

Data Type

```python
print(arr.dtype)
```

Size

```python
print(arr.size)
```

Zeros

```python
a = np.zeros((3,3))
print(a)
```

ones

```python
b = np.ones((2,4))
print(b)
```

Sequence

```python
c = np.arange(0,10,2)
print(c)
```

Random Module

```python
import numpy as np

arr = np.random.randint(1,10,(3,3))
print(arr)
```

Indexing

```python
print(arr[0,1])
```

Slicing

```python
print(arr[0:2,1:3])
```

Array Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a+b)
print(a*b)
```

Broadcasting

```python
a = np.array([[1,2,3],
              [4,5,6]])

b = np.array([10,20,30])

print(a + b)
```

Trignometric Functions

```python
import numpy as np

angles = np.array([0, np.pi/2, np.pi])

print(np.sin(angles))
print(np.cos(angles))
print(np.tan(angles))
```

Statistical Functions

```python
arr = np.array([10,20,30,40,50])
```

Mean

```python
print(np.mean(arr))
```

Median

```python
print(np.median(arr))
```

Standard Deviation

```python
print(np.std(arr))
```

Variance

```python
print(np.var(arr))
```

Min/Max

```python
print(np.min(arr))
print(np.max(arr))
```

Sum

```python
print(np.sum(arr))
```

Shape Manipulation
Reshape

```python
arr = np.arange(1,7)

new_arr = arr.reshape(2,3)

print(new_arr)
```

flatten

```python
print(new_arr.flatten())
```

Ravel

```python
print(new_arr.ravel())
```

Transpose

```python
print(new_arr.T)
```

Joining Arrays

```python
a = np.array([[1,2]])
b = np.array([[3,4]])

print(np.concatenate((a,b)))
```

hstack

```python
print(np.hstack((a,b)))
```

Splitting Arrays

```python
arr = np.array([1,2,3,4,5,6])

print(np.split(arr,3))
```

Linear Algebra

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])
```

Matrix Multiplication

```python
print(np.dot(A,B))
```

Inverse

```python
print(np.linalg.inv(A))
```
