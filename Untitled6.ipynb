Indexing

```python
import numpy as np

arr = np.array([[10, 20, 30],
                [40, 50, 60],
                [70, 80, 90]])
print(arr[1, 2])
print(arr[1])
print(arr[:, 1])
print(arr[0:2, 1:3])
```

Stacking

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])

v = np.vstack((a, b))
print(v)

h = np.hstack((a, b))
print(h)
```

Column Stack

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

col = np.column_stack((x, y))
print(col)
```

Memory Optimization (32-bit vs 64-bit)

```python
# Default (usually int64)
a = np.array([1, 2, 3])
print(a.dtype)

# Optimized (int32)
b = np.array([1, 2, 3], dtype=np.int32)
print(b.dtype)

# Float optimization
c = np.array([1.0, 2.0], dtype=np.float32)
print(c.dtype)
```

Avoid Unnecessary Copies

```python
arr = np.array([1, 2, 3, 4])

# View (no copy)
view = arr[1:3]
view[0] = 99
print(arr)  # original changes!

# Copy (safe)
copy = arr[1:3].copy()
copy[0] = 0
print(arr)  # original unchanged
```

Strides

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr.strides)
```

Contiguous Arrays

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr.flags['C_CONTIGUOUS'])  # True


t = arr.T
print(t.flags['C_CONTIGUOUS'])  # False


t2 = np.ascontiguousarray(t)
print(t2.flags['C_CONTIGUOUS'])  # True
```

Iteration Techniques(Basic Iteration)

```python
arr = np.array([[1, 2], [3, 4]])

for row in arr:
    print(row)
```

Iteration Techniques(Element-wise Iteration)

```python
for val in np.nditer(arr):
    print(val)
```

With Index

```python
for index, value in np.ndenumerate(arr):
    print(index, value)
```
