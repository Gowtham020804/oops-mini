Handling Missing Values

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "A": [1, 2, np.nan, 4],
    "B": [5, np.nan, np.nan, 8]
})
print(df.isnull())
print(df.isnull().sum())
df_drop = df.dropna()
df_fill = df.fillna(0)
df["A"] = df["A"].fillna(df["A"].mean())
df_ffill = df.fillna(method="ffill")
```

Removing Duplicates

```python
df = pd.DataFrame({
    "Name": ["A", "B", "A", "C"],
    "Age": [20, 25, 20, 30]
})
print(df.duplicated())
df_clean = df.drop_duplicates()
df_clean = df.drop_duplicates(subset=["Name"])
```

Handling Outliers

```python
df = pd.DataFrame({
    "Salary": [20000, 25000, 30000, 1000000]  # outlier
})

Q1 = df["Salary"].quantile(0.25)
Q3 = df["Salary"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Filter out outliers
df_clean = df[(df["Salary"] >= lower) & (df["Salary"] <= upper)]
```

String Cleaning

```python
df = pd.DataFrame({
    "Name": [" Alice ", "BOB", "charlie"]
})
df["Name"] = df["Name"].str.strip()
df["Name"] = df["Name"].str.lower()
df["Name"] = df["Name"].str.upper()
df["Name"] = df["Name"].str.replace("BOB", "Robert")
```

Renaming Columns



```python
df = pd.DataFrame({
    "First Name": ["A", "B"]
})
df = df.rename(columns={"First Name": "first_name"})
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

Handling Invalid Data

```python
df = pd.DataFrame({
    "Age": [20, -5, 30]
})
df = df[df["Age"] > 0]
```

Handling Categorical Data

```python
df = pd.DataFrame({
    "Gender": ["Male", "Female", "Male"]
})

# Encode categories
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
```

Scaling

```python
from sklearn.preprocessing import MinMaxScaler

df = pd.DataFrame({
    "Salary": [20000, 30000, 40000]
})
scaler = MinMaxScaler()
df["Salary"] = scaler.fit_transform(df[["Salary"]])
```
