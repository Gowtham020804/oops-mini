Load Dataset

```python
import pandas as pd

df = pd.read_csv("train.csv")

print(df.head())
```

Dataset Info

```python
print(df.info())
```

Shape of Dataset

```python
print(df.shape)
```

Selecting Important Columns

```python
print(df[["Name", "Age", "Sex", "Fare"]].head())
```

Sorting Data(Sort by age)

```python
print(df.sort_values(by="Age"))
```

Sort by Fare Descending

```python
print(df.sort_values(by="Fare", ascending=False))
```

Filtering Data(Female passengers)

```python
print(df[df["Sex"] == "female"])
```

First-class passengers

```python
print(df[df["Pclass"] == 1])
```

Grouping Operations(Group by Gender)

```python
group_gender = df.groupby("Sex")
print(group_gender["Survived"].mean())
```

Group by Passenger Class

```python
group_class = df.groupby("Pclass")
print(group_class["Fare"].mean())
```

Multiple Grouping

```python
group_multi = df.groupby(["Sex", "Pclass"])
print(group_multi["Survived"].mean())
```

Aggregation Functions

```python
print(df.groupby("Pclass")["Age"].agg(["mean", "min", "max"]))
```

Value Counts(Count Male/Female)

```python
print(df["Sex"].value_counts())
```

Count Survival

```python
print(df["Survived"].value_counts())
```

Survival based on Class

```python
print(pd.crosstab(df["Pclass"], df["Survived"]))
```

Missing Value Analysis

```python
print(df.isnull().sum())
```

Apply Function
Age Category

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["Age_Group"] = df["Age"].fillna(0).apply(age_group)

print(df[["Age", "Age_Group"]].head())
```
