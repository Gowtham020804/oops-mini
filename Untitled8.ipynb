Load external dataset

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Dataset Loaded Successfully")
```

View First & Last Rows

```python
# First 5 rows
print("HEAD:")
print(df.head())

# Last 5 rows
print("\nTAIL:")
print(df.tail())
```

Filtering Data

```python
# Filter flowers with sepal_length > 5.0
filtered_df = df[df["sepal_length"] > 5.0]

print(filtered_df.head())
```

Multiple conditions

```python
# sepal_length > 5 AND species is setosa
filtered_df2 = df[(df["sepal_length"] > 5) & (df["species"] == "setosa")]

print(filtered_df2)
```

Select specific columns after filtering

```python
filtered_df3 = df[df["petal_length"] > 4][["sepal_length", "species"]]

print(filtered_df3.head())
```

Combine Filtering + Head/Tail

```python
# First 3 rows after filtering
print(df[df["petal_width"] > 1.5].head(3))

# Last 3 rows after filtering
print(df[df["petal_width"] > 1.5].tail(3))
```
